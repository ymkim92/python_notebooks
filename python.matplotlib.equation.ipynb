# Python으로 방정식 그래프 그리기

Python을 이용하면 수학시간에 배운 방정식을 쉽게 그래프로 그릴 수 있다.
본 문서에서는 간단한 2차 방정식 y = x**2 의 그래프를 그리는 방법을 예제로 설명한다.
먼저 list를 이용해서 그리는 방법을 살펴본 뒤, numpy로 그리는 방법을 볼 것이다.

## list를 이용한 방법


```python
import matplotlib.pyplot as plt

x = [i for i in range(-5, 6)]
y = [i**2 for i in range(-5, 6)]
plt.plot(x, y)
plt.show()
```

## numpy를 이용한 방법


```python
import matplotlib.pyplot as plt
import numpy as np

# x = np.array(range(-5, 6)) # same as below
x = np.arange(-5, 6)
y = x ** 2
plt.plot(x, y)
plt.show()
```

numpy를 이용할 경우 방정식을 `y = x ** 2` 와 같이 수학식의 표현 그대로 쓸 수 있어 가독성이 좋다.
`arange()`는 옵션으로 간격을 명시할 수 있고, `linspace()`는 샘플의 수를 옵션으로 사용한다.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-5, 5, 11)
y = x ** 2
plt.plot(x, y)
plt.show()
```

샘플수를 50으로 늘리면 부드러운 곡선을 그릴 수 있다.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-5, 5, 50)
y = x ** 2
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('y=x**2')
plt.show()
```
